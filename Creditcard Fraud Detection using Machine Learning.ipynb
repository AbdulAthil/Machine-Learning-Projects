## Split data into X & y

```python
X=df.drop(["Class"],axis=1)
X
```

```python
y=df["Class"]
y
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaler = sc.fit_transform(X)
```

```python
x_scaler
```

## Split data into test & train

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaler,y,test_size=0.25,random_state=1)
```

## Model Training

```python
#classify function
from sklearn import metrics
from sklearn.model_selection import cross_val_score

def classify(model,x,y):
    x_train,x_test,y_train,y_test=train_test_split(x_scaler,y,test_size=0.25,random_state=1)
    model.fit(x_train,y_train)
    
    prediction=model.predict(x_test)
    print("Accuracy is",model.score(x_test,y_test)*100)
    score=cross_val_score(model,x,y,cv=5)
    print("Cross Validation Score is",np.mean(score)*100)
    report=metrics.classification_report(y_test,prediction)
    print("Classification Report:\n",report)
    matrix=metrics.confusion_matrix(y_test,prediction)
    print("Confusion Matrix:\n",matrix)
    accuracy=metrics.accuracy_score(y_test,prediction)
    print("Accuracy Score:",accuracy*100)
    plt.figure(figsize=(4,3))
    sns.heatmap(matrix,annot=True)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
classify(model,x_scaler,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
classify(model,x_scaler,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model= KNeighborsClassifier(n_neighbors=5)
classify(model,x_scaler,y)
```

```python
from sklearn.svm import SVC
model=SVC()
classify(model,x_scaler,y)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
classify(model,x_scaler,y)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
classify(model, x_scaler, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
classify(model, x_scaler, y)
```

As we can see,the ExtraTreesClassifier algorithm gives us more accuracy than the all other algorithms.
