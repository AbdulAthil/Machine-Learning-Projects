## Split the data into train and test

```python
df.head()
```

```python
x=df.drop(columns=["Outcome"],axis=1)
x
```

```python
y=df["Outcome"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
print("x_train:",x_train.shape)
print("x_test:",x_test.shape)
print("y_train:",y_train.shape)
print("y_test:",y_test.shape)
```

## Model Building

```python
from sklearn.model_selection import cross_val_score
from sklearn import metrics
```

```python
models=[]
scores=[]

def classify(model,x,y):
    model.fit(x_train,y_train)
    #prediction
    y_predict=model.predict(x_test)
    #Evaluation
    #Accuracy
    Accuracy=model.score(x_test,y_test)*100
    print("Accuracy is:",Accuracy,"\n")
    #Cross Validation
    score=cross_val_score(model,x,y,cv=5)
    print("Cross validation score:",np.mean(score)*100,"\n")
    #Classification Report
    report=metrics.classification_report(y_test,y_predict)
    print("Classification Report:\n",report,"\n")
    #Confusion Matrix
    matrix=metrics.confusion_matrix(y_test,y_predict)
    print("Confusion Matrix:\n",matrix,"\n")
    #Heatmap for confusion matrix
    plt.figure(figsize=(4,3))
    sns.heatmap(matrix,annot=True)
    plt.show()
    
    models.append(model)
    scores.append(Accuracy)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
classify(model,x,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#Error Rate
error_rate=[]
for i in range(1,20):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    y_predict=model.predict(x_test)
    error_rate.append(np.mean(y_predict!=y_test))
```

```python
#Error rate vs K-values

plt.figure(figsize=(10,6))
plt.plot(range(1,20),error_rate,color="blue",marker="o",mfc="red",markersize=10)
plt.title("Error Rate vs K-values")
plt.xlabel("K-values")
plt.ylabel("Error Rate")
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=13)
classify(model,x,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
classify(model,x,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100)
classify(model,x,y)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model=ExtraTreesClassifier()
classify(model,x,y)
```

```python
from sklearn.svm import SVC
model=SVC(kernel="linear")
classify(model,x,y)
```

```python
from xgboost import XGBClassifier
model=XGBClassifier()
classify(model,x,y)
```

```python
print(models)
print(scores)
```

```python
plt.figure(figsize=(10,6))
a=[1,2,3,4,5,6,7]
colors=["Skyblue","pink","grey","cyan","lightgreen","yellow","purple"]
plt.bar(a,scores,color=colors,width=0.5)
plt.plot(a,scores,lw=0.5,color="green",marker="o",ms=8,mfc="red")
plt.title("Model Comparision")
print(f'{models[0]} : ',scores[0])
print(f'{models[1]} : ',scores[1])
print(f'{models[2]} : ',scores[2])
print(f'{models[3]} : ',scores[3])
print(f'{models[4]} : ',scores[4])
print(f'{models[5]} : ',scores[5])
print("XBGClassifier : ",scores[6])
```

From the above comparision we can see that the RandomForestClassifier gives us more accuracy than the other models

## Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
param_grid={"n_estimators":[10,50,100,150,200]}

rf_classifier = RandomForestClassifier()

grid_search=GridSearchCV(estimator=rf_classifier,param_grid=param_grid,cv=5)
grid_search.fit(x_train,y_train)

best_n_estimators=grid_search.best_params_["n_estimators"]
print("best_n_estimator : ",best_n_estimators)

y_pred=grid_search.predict(x_test)

#Accuracy
accuracy=grid_search.score(x_test,y_test)
print("Accuracy is : ",accuracy*100)
#classification report
report=metrics.classification_report(y_test,y_pred)
print("Classification Report :\n\n",report)
#Confusion Matrix
matrix=metrics.confusion_matrix(y_test,y_pred)
print("Confusion Matrix:\n",matrix)
```

After hyperparameter Tuning,we can see that the accuracy of RandomForestClassifier model slightly increases.
